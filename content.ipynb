```python
def get_article_content():
    
    # url definition
    url = "https://www.autopilotreview.com/cars-with-autopilot-self-driving/"
    
    # Request
    r = requests.get(url)
    r.status_code

    # We'll save in coverpage the cover page content
    coverpage = r.content

    # Soup creation
    soup = BeautifulSoup(coverpage, 'html5lib')

    # News identification
    article_text = soup.find_all('a', class_='tdb-logo-a')
    len(article_text)
    
    number_of_articles = 1

    # Empty lists for content, links and titles
    news_contents = []
    list_links = []
    list_titles = []

    for n in np.arange(0, number_of_articles):

        # Getting the link of the article
        link = article_text[n]['href']
        list_links.append(link)

        # Getting the title
        title = article_text[n].get_text()
        list_titles.append(title)

        # Reading the content (it is divided in paragraphs)
        article = requests.get(link)
        article_content = article.content
        soup_article = BeautifulSoup(article_content, 'html5lib')
        body = soup_article.find_all('div', class_='tdb-logo-a')
        x = soup_article.find_all('p')

        # Unifying the paragraphs
        list_paragraphs = []
        for p in np.arange(0, len(x)):
            paragraph = x[p].get_text()
            list_paragraphs.append(paragraph)
            final_article = " ".join(list_paragraphs)

        news_contents.append(final_article)

    # df_features
    df_features = pd.DataFrame(
         {'Content': news_contents 
        })

    # df_show_info
    df_show_info = pd.DataFrame(
        {'Article content': list_titles,
         'Article Link': list_links})
    
    return (df_features, df_show_info)
```

```python
start = time.time()
x, y = get_article_content()
end =time.time()
te = end-start
print("The time elapsed is %f seconds" %(te))
```
